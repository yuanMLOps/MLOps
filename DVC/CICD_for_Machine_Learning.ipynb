### Introduction to YAML
#### What is YAML?
* YAML is a a Markup Language (yet another markup language)
* It is used in configuration files, data exchange and structured data representation
* It is a data formatting language similar to JSON and XML
* Allows a standard format to transfer data between languages or applications for interoperability and seamless communication
* The vaid file extensions include ymal or yml
* It is used to write configuration for variety of CI/CD tools such as GitHub Actions and Data Version Control (DVC)rol (DVC)


#### YAML Syntax
* YAML has a hierarchical structure defined by indentation
  + tab is not allowed
  + comments start with #
* YMAL Scalars
  + Numbers: integers or floating-point numbers
  + Booleans: true or false
  + Null: keyword null or ~
  + Strings: can be plain text or enclosed in '' or ""
* YMAL Collections
  + Sequences (also called list, arrays or vectors) written in two styles
    + block style
    + flow stype
      ```YAML
      # block style list
      - first
      - second
      - third
      # flow style list
      [first, second, third]
      ```
  + Mappings: keys and value pairs separated by :
    ```YAML
    key1: value1
    blocklist:
      - first
      - second
    flowlist: [1.2, 2, "fifty", true]
    ```

#### Intermediate YAML
* Multiline strings: Block scalar format
  + represent multiline strings or block of text
  + used for
    + Code snippets (e.g. shell commands)
    + paragraphs (e.g. logs)
    + configuration files
  + style indicators - literal
    + this doesn't work on the last line, which is governed by chomping style indicators
    + literal style (|): this preserves line break and indentation
    + useful for writing shell commands
```YAML
commands: |
  echo "Running python script..."
  python3 script.py
```
  + style indicators - fold
    + fold style (>): removes line breaks into spaces
    + useful for writing wrapped messages like paragraphs
    + line breaks followed by a blank line are preserved as line breaks
    + lines with extra indentation are also not folded
  + style indicators - chomping style indicator: clip
    + clip: default mode for (|), which adds a single newline at the end
  + style indicators - chomping style indicator: keep
    + keep style (+): retains all newlines at the end of the string
    + added after the style indicators. e.g.
      + use |+ to reserve multiline in the text and keep new lines in the end
      + use >+ to remove line breaks (except breaks followed by blan line) and keep new lines in the end
  + style indicators - chomping style indicator: strip
    + strip style (-): removes all newlines at the end of the string
    + added after the style indicators. e.g.
      + use |- to reserve multiline in the text and remove new lines in the end
      + use >- to remove line breaks (except breaks followed by blan line) and remove new lines in the end
* Dynamic value injection (not supported by all YAML parsers)
  + expressions allow parsers to dynamically substitute values
  + usages:
    + environment variables
    + references to other part of YAML
  + typically represented by
```YAML
${{}}
```
  + example:
```YMAL
database:
  host: ${{ config.database.host }}
  port: ${{ config.database.port }}
```

* multi-document YAML
  + use `---` to separate multiple YAML content so that we can write multipl YMAL file content in one file

### Introduction to Github Actions
#### What is Github Actionss
* GitHub Actions (GHA): CI/CD platform to automate pipelines
* Pipeline: a sequence of interconnected steps/stages that represent the flow of work and data
* General workflow of GHA:
  + checkout -> enviornment set up -> lint & test -> build -> deploy
* GHA components
  + Event: a specific activity in a repository that triggers a workflow run
    + Push
    + Pull Request
    + Opening an issue
  + Workflow: automated process that will run one or more jobs
    + defined in YAML file
    + Triggered automatically by event
      + Manual run possible
    + Housed in .github/workflows directory in the repository
    + Multiple workflows can be defined
  + Steps and actions: individual unit of work
    + Executed in order, depends on previous step
    + run on the same machine, so data can be shared
    + unit of work example
      + compiled code application, shell script
      + Action: GHA platform specific application (you can consider it as libraries, or sharedlibrary in Jenkins)
        + E.g. checkout repo, comment on PR
  + Jobs and Runners
    + each job is independent as a set of steps (unlike steps, which depend on other steps)
    + parallel execution is possible
    + all steps of the same job are executed on the same compute machine called runners
* A simple GHA workflow:
  + Commit push -> checkout Repo -> Run Python code -> stop
  + event: Push
  + job: runs on Ubuntu runner, has two steps
    + Action: checkout Repo
    + run python code

#### Setting a basic CI pipeline
* Anatomy of GitHub Actions workflow
```YAML
# Name of the workflow
name: CI

# Events that trigger workflow
on:
  push: # event
    braches: ["main"]

# Define jobs and runners
jobs:
  build: # the name of the first job and its definition
    runs-on: ubuntu-latest
# Define steps
    steps:
      - name: Run a multi-line script
        run: |
          echo Hello, world!
          echo Add other actions to build
          echo test, and deploy your project
```

* configure github repo for GHA in github repo
  + go to settings/actions ->General, set "Allow all actions and reusable workflows" in "Actions permissions"
  + go to "Actions" -> Automation/Manual workflow -> define <your_workflow.yaml> file in /.github/workflows/ -> commit
  + Inspect workflow run by clicking "Actions" tab again. You will see your workflow. Click the workflow. you will see your jobs execution status
  + click the specific job, you will see a page showing the logs of the job
* Workflow for Pull Requst
  + multiple collaborators work on same repo (if you just update code on main by a few developer, use push/poll is enough)
  + feature/topic branches are created for related work. Once are done
    + pull requests are opened
    + code is reviewed and changes are suggested
    + branch code is merged into main branch
  + CI/CD enabled checks improve code quality
    + in workflow yaml file, set event to pull_request
```YAML
on:
  pull_request:
    branches: ["main"]
```
* Using Actions in steps
  + Actions are functions defined in another github repo
  + you can pass the function parameters by defining action-argument: value using with clause
```YAML
steps:
  - name: My Action
    uses: <git_url>@<version_number>
    with:
      action-argument: value
```

* Config a typical workflow steps using existing actions provided by github
  + step 1: check out repo
  + step 2: set up python environment
```YAML
name: PR

on:
  pull_request:
    branches: ["main"]

jobs:
  build:
    runs-on: ubuntu-latest    
    steps:
      - name: Checkout
        uses: actions/checkout@v3    
      - name: Set up Python
        uses: actions/setup-python@v4
        with:
          python-version: 3.9
      - name: Run Python script
        run: |
          echo hello_world.py
          python hellow_world.py
```      

#### Contexts in GHA
* contexts are predefined environment variables or data that can be accessed by workflow runs 
  + worflow runs (using github contexts)
  + variables
    + env context for non-sensitive set in workflow
    + secret context for sensitive information set in GHA GUI
  + runner environments (using runner contexts)
  + current jobs (using job contexts)
  + access contexts using the expression syntax: `${{ context.XXX }}`
* Use variables by env
  + defined as a 'key: value' pair in plain text under 'env' in YAML file
  + you can define 'env' entry at the level you need (global, under jobs, steps etc.)
  + referenced by `${{ env.<variable_name> }}
  + Global/local scope is controlled by the level where defined
  + code example:
```YAML
name: Greeting on variable day

# global env
env:
  Greeting: Hello
jobs:
  greeting_job:
    runs-on: ubuntu-latest

    # local env: scoped to greeting_job
    env:
      First_Name: Yuan
    steps:
      - run: |
          echo "${{ env.Greeting }} \
          ${{ env.First_Name }}."
```

* Use Secrets
  + store sensitive information in encrypted manner
    + passwords, API keys
  + access values via secrets context `${{ secrets.SuperSecret }}
  + use secrets either as an environment variable under env key, or as argument to an action using 'with' key
  + Define them as input or environment variables to retrieve and use in workflow
  + need to set up in github repo
    + go to github repo -> settings -> Secrets and variables -> Actions -> secrets/variables
  + code example
```YAML
steps:
  - name: Hello world action
    env: # we can set the secret as an env variable to retrieve
      super_secret: ${{ secrets.SuperSecret }}
    with: # we pass secret value as an input to an action 
      super_secret: ${{ secrets.SuperSecrete }}
  - name: Print Secret
    run: |
      echo "my secret is \
      ${{ secrets.SuperSecret }}"
```

* GITHUB_TOKEN secret
  + built in secret provided by GHA
  + used to perform workflow actions
    + cloning the repo and fetching code
    + opening and closing issues and pull requests
  + Automatically available in every GHA workflow
    + accessed by `${{ secrets.GITHUB_TOKEN }}`
  + permissions can be tuned to the right degree
  + code example
```YAML
permissions:
  pull-requests: write
steps:
  - name: Comment PR
    uses: thollander/actions-comment-pull-request@v2
    with:
      GITHUB_TOKEN: ${{ secrets.GITHUB_TOKEN }}
      message: |
        Hello world ! :wave:
```        


          
  

### Continuous Integration (CI) in Machine Learning
#### Model training with GitHub Actions
* A Modeling workflow can consist of the following stages:
  + data preprocessing
    + convert categorical features to numerical
      + target encoding
        + convert categorical variables to numerical ones w/o significant complexity
        + when large dimensionality of categorical features prevents us from using one hot encoding
        + we encode each categorical variable by its corresponding average value in target column
    + replace missing values of features
    + scale features
  + model training
    + fit data to a random forest classifier
      + hyperparameter tuning: `max_depth=2, n_estimators=50`
  + model evaluation on test dataset
    + performance plots
      + confusion matrix plot
* Contiuous Machine Learning (CML)
  + an open source tool for implementing CI/CD in ML
  + as a GHA action, CML is used for machine provisioning, model training and evaluation, comparing ML experiments and monitoring changing datasets.
  + can help train and evaluate models and then generate a visual report with results and metrics automatically on every pull request.
  + How to set up CML in workflow yaml file
    + set up your ML workflow
      + develop data_preprocessing.py module to output the preprocessed dataset
      + develop train.py module to generate train/test datasets, train model and evaluate model by test dataset
      + make sure the model performance metrics are saved in a json file, and plots are generated in .png files
    + crate a CML job under jobs with GIHUB_TOKEN permissions of write
    + define enbale CML to use CML action
    + call train.py file as you would do in command line. CML will trace the training
    + add resutls.txt and performace plot to report.md
    + generate cml comment from report.md
  + code example:
```YAML
name: CML model training workflow
on:
  pull_request:
    branches: main
  permissions: write-all 
  
jobs:
  train_and_report_eval_performance:
    runs-on: ubuntu-latest
    permissions:
      pull-requests: write-all
    steps:
      - name: checkout
        uses: actions/checkout@v3

      - name: setup python
        uses: actions/setup-python@v4
        with:
          python-version: 3.9
          
      - name: setup GHA CML action
        uses: iterative/setup-cml@v1
        
      - name: Train model
        run: |
          # your ML workflow goes here
          pip install -r requirements.txt
          pyton3 preprocess_dataset.py 
          python3 train.py
          
      - name: Write CML report        
        env: # local env: scoped to greeting_job
          REPO_TOKEN: ${{ secrets.GITHUB_TOKEN }}
        run: |
            # Add results and plots to markdown
            cat results.txt >> report.md
            echo "![training graph](./graph.png)" >> report.md

            # create comment from markdown report
            cml comment create report.md       
```          

#### Versioning datasets with Data Version Control
* versioning ensures a hisotrical record of data changes
* data versioning is critical
  + reproducibility: e.g. replicate model training performed by others
  + experimentation and iteration: Track lineage of model performance across architectures and training datasets
  + Collaboration
  + bug tracking and debugging
  + model monitoring and maintenance
  + auditing and validation such as in healthcare or finance
* Data Version Control (DVC)
  + manages data and experiments
  + silimar to git and integrates with Git
  + git tracks metadata and code, DVC tracks data so that we can manage both
* DVC storage
  + maintain records of various data versions using Git and store the data separately using remote storage
  + data storage can be implemented locally by SSH, HTTP/HTTPS, Local file system or cloud object storage
* setup DVC
  + install dvc by `pip install dvc`
  + initialize git if not by `git init`
  + initialize dvc if not by `dvc init`
    + after this, dvc generates the following files in .dvc folder (.dvc file is outside of this folder):
      + .gitignore: dvc-specific files that shuldn't be tracked by git
      + config: configuration setting for the dvc project, such as remote storage location, compression settings
      + tmp: contains caches, logs, or other temporary data created when interacting with DVC
  + track data files by `dvc add data.csv`
    + a .dvc file as a placeholder containing file metadata is generated
    + each dvc tracked data file has its corresponding .dvc file (data.csv -> data.csv.dvc)
    + checked into git to manage data versions (only .dvc file, not original data file)
    + the dvc cache, a hiddent storage for files and directories tracked by DVC, is populated at .dvc/cache
    + data.csv.dvc contains metadata and checksum of data file, specifically,
      + the md5 of the data file to track if the content of data file changes
      + size: file size
      + hash: md5
      + path: path to the data file in the project folder

#### Interacting with DVC remotes
* what is dvc remotes
  + Git is good at software version control, but services like github that hosts git repos, imposes storage limits
  + since github/bitbucket can't be used to track data, DVC use external storage (DVC remote) for this
  + similar to git remotes but for cached data
* why use DVC?
  + synchronize large files and directories monitored by DVC
  + centralized data repo to facilitate sharing and collaboration
  + archiving diverse iterations of datasets and models, thereby conserving local storage capacity
* implementation of DVC remote
  + supports a variety of cloud-based providers such as AWS, Azure and GCP
  + on-prem local storage accessible via SSH and HTTP
  + set remotes using `dvc remote add myAWSremote s3://mybucket`, which is saved in .dvc/config
  + we can modify the remote settings by `dvc remote modify`, e.g `dvc remote modify myAWSremote connect_timeout 300`
  + you can also add dvc remote using a local foler by `dvc remote add --local`
  + you can set up default remote by `dvc remote add -d` command, e.g. `dvc remote add -d mylocalremote /tmp/dvc`, which will be saved in core section of .dvc/config
  + all these settings are stored in .dvc/config file as the following:
  + ```text
    [core]
    remote = mylocalremote
    ['remote "myAWSremote"]
      url=s3://mybucket
      connect_timeout = 300
    ``` 
* commands to transfer data from local workspace to remote storage locations
  + `dvc push <target>` or `dvc push` push to remote
  + `dvc pull <target>` or `dvc pull` pull from remote
  + .dvc is tracked by git, not dvc
* tracking data changes
  + first, change data file contents
  + add dataset changes in dvc to track the change by `dvc add dataset.csv`. this will
    + change the dataset.csv.dvc file
    + add dataset.csv to .gitignore file if not
  + commit changed .dvc file to git and changed data file to dvc remote
    ```bash
    # update metadata tracking by git
    git add dataset.csv.dvc
    git commit dataset.csv.dvc -m "dataset updates"
    git push origin main

    # update data file to dvc remote
    dvc push
    ```
#### DVC Pipeline
* why need DVC Pipeline
  + versioning data files with DVC alone is not very useful because
    + for ML use cases, data needs to be filtered, cleaned, and tranformed before model training
    + there is no need to repeat each step than don't introduce change
    + ML processes depend on each other and can be expressed as Direcred Acyclic Graph (DAG)
* what is a DVC pipeline
  + a structured sequence of stages that defines the workflow and its dependencies for a ML or data processing task
  + these steps are defined in dvc.yaml file including
    + input data and scripts (dep)
    + stage execution commands (cmd)
    + output artifacts (outs)
      + can include special data with their respective keys e.g. metrics and plots
  + similar to GHA workflow but
    + focused on ML tasks instead of CI/CD
    + we can run DVC pipeline inside GHA workflow
* how to create DVC pipleline (dvc.yaml)
  + use `dvc stage add` command to generate dvc.yaml
  + we can use multiple -d or -o for multiple dependencies and outputs
  + run pipeline by `dvc repro`
    + a dvc.lock file is generated
    + similar to .dvc file, dvc.lock file captures md5 hashes for each stage
    + it is the best practice to commit to git immediately after dvc.lock is created or modified by
      `git add dvc.lock && git commit -m "first pipeline repro"`
```YAML
dvc stage add \
-n preprocess \
-d raw_data.csv -d preprocess.py \
-o processed_data.csv \
python preprocess.py
```
* execution of DVC pipeline
  + if we don't change the dependencies of a step, it will skip its execution when rerunning the pipeline
  + this is particularly useful in complex DAGs to skip unchanged steps
  + we can use dvc dag command for pipeline visualization
