### Ubuntu commands
* collection of some useful linux commands on Ubuntu when setting up system

#### Extend disk space to be used by K8s cluster and longhorn
```bash

sudo pvresize /dev/sda3 # tells LVM that /dev/sda3 is larger now
sudo vgs # verify free space exists in ubuntu-vg
sudo  lvextend -r -l +100%FREE /dev/ubuntu-vg/ubuntu-lv # extend the logical volume and filesystem

df -h / # expected result: /dev/mapper/ubuntu--vg-ubuntu--lv ~85G (show the extended space avaliable
```
